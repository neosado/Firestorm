```julia
using PyPlot

l = 0.2
N = 100

x = zeros(N)
x[1] = 1.1
for i = 2:N
    x[i] = exp(l * (x[i-1] - 1))
end

plot(1:N, x)

;
```

```julia
using PyPlot

dist = 0:10

plot(dist, exp(-dist/20))
legend(["e", "2"])

n = 4
dist = norm([n, n] - [1, 1])
println(dist)
println(0.5 + 0.5 * exp(-dist/20))

;
```

```julia
using Distributions, PyPlot

n = 1000000
mu = 10.
sigma = 1.

x = zeros(Int64, n)
for i = 1:n
    x[i] = int(rand(TruncatedNormal(mu, sigma, 1, Inf)))
end

PyPlot.hist(x, 100)

;
```
