```julia
using Reactive, Interact
using DataStructures

include("ScenarioOneIJUtil.jl");
```

```julia
f = figure(facecolor = "white")
@manipulate for param_set = togglebuttons(OrderedDict(["1", "2"], [1, 2]), label = "param set"), policy = [:back, :landing, :stay], uncertainty = slider(0.:1.:2., value = 0., label = "uncertainty"); withfig(f) do
        plotEvaluation(param_set, policy, uncertainty, fig = f)
    end
end
```

```julia
f = figure(facecolor = "white")
@manipulate for param_set = togglebuttons(OrderedDict(["1", "2"], [1, 2]), label = "param set"), uncertainty = slider(0.:1.:2., value = 0., label = "uncertainty"); withfig(f) do
        plotPolicy(param_set, uncertainty, fig = f)
    end
end
```

```julia
param_set = 1
policy = :stay

for uncertainty in [0., 1., 2.]
    plotEvaluation(param_set, policy, uncertainty);
    savefig("eval_" * string(param_set) * "_" * string(policy) * "_" * string(int(uncertainty)) * ".png")
end
```
