```julia
using Reactive, Interact

include("ScenarioOneIJUtil.jl");
```

```julia
f = figure(facecolor = "white")
@manipulate for policy = [:back, :landing, :stay], uncertainty = slider(0.:1.:2., value = 0., label = "uncertainty"); withfig(f) do
        plotEvaluation(policy, uncertainty, fig = f)
    end
end
```

```julia
f = figure(facecolor = "white")
@manipulate for uncertainty = slider(0.:1.:2., value = 0., label = "uncertainty"); withfig(f) do
        plotPolicy(uncertainty, fig = f)
    end
end
```

```julia
policy = :stay

for uncertainty in [0., 1., 2.]
    plotEvaluation(policy, uncertainty);
    savefig("eval_" * string(policy) * "_" * string(int(uncertainty)) * ".png")
end
```

```julia
for uncertainty in [0., 1., 2.]
    plotPolicy(uncertainty);
    savefig("policy_" * string(int(uncertainty)) * ".png")
end
```
