```julia
using Reactive, Interact
using DataStructures

include("ScenarioOneIJUtil.jl")

database = "s1results.jld";
```

```julia
f = figure(facecolor = "white")
@manipulate for param_set = togglebuttons(OrderedDict(["1", "2"], [1, 2]), label = "param set"), policy = [:back, :landing, :stay], uncertainty = slider(0.:1.:2., value = 0., label = "uncertainty"); withfig(f) do
        plotEvaluation(param_set, policy, uncertainty, fig = f, database = database)
    end
end
```
